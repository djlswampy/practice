```python
# OpenAI client를 통해 llm 호출

from openai import OpenAI

client = OpenAI()

response = client.responses.create(
  model="gpt-5-nano",
  input="Tell me a three sentence bedtime story about a unicorn in korean"
)

print(response)
```
